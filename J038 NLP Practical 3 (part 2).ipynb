```python
import pandas as pd
import re

def extract_and_format_date(text):
   
    patterns = [
        # Pattern for dates in format YYYY-MM-DD
        (r'\b(\d{4})-(\d{2})-(\d{2})\b', r'\3/\2/\1'),
        # Pattern for dates in format DD/MM/YYYY
        (r'\b(\d{2})/(\d{2})/(\d{4})\b', r'\1/\2/\3'),
        # Pattern for dates in format DD Month YYYY 
        (r'\b(\d{1,2})(?:st|nd|rd|th)? (\w+),? (\d{4})\b', None),
        # Pattern for dates in format Month DD, YYYY 
        (r'\b(\w+) (\d{1,2}), (\d{4})\b', None)
    ]

    month_dict = {
        'January': '01', 
        'February': '02',
        'March': '03', 
        'April': '04',
        'May': '05', 
        'June': '06', 
        'July': '07', 
        'August': '08',
        'September': '09', 
        'October': '10', 
        'November': '11', 
        'December': '12'
    }

    for pattern, replacement in patterns:
        match = re.search(pattern, text)
        if match:
            groups = match.groups()
            if replacement:
                return re.sub(pattern, replacement, text)
            else:
                day, month_name, year = groups
                month_number = month_dict.get(month_name, '00')  # Default to '00' if month not found
                return f'{day.zfill(2)}/{month_number}/{year}'
    
    return 'Date not found'

```

```python
df['Parsed Date'] = df['Input'].apply(extract_and_format_date)

df
```
